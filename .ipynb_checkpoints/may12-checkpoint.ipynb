# Non-Maximum Suppression

Objective:   
Unambiguous definition of edges with single pixels, edge tapering.

>- calculate gradients and angle of gradients (azimuths of gradients)
>- selecting pixels in the edge with the highest values 

## Gradients with sobel kernel
>- kx,ky - kernel
>- `gx,gy` - gradients
>- $grad = \sqrt{gx^2 + gy^2}$ - gradient


```python
ky = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
kx = ky.T
kx,ky
```

```python
# How does this work?
tmp=blur[200:203,200:203]
print(f'tmp:\n{tmp}\n\nkx * tmp:\n{kx * tmp}\n\nres:\n{(kx * tmp).sum()}')
```

```python
gx = ndimage.convolve(blur,kx)
gy = ndimage.convolve(blur,ky)
grad = np.hypot(gx,gy)

mymod.myshow([gx,gy,grad],1,3,5)
```

## Gradien azimuth `gradAz`:
>- $gradAz$ - angle (azimuth) of the gradient 
>- $\tan(gradAz) = \frac{gy}{gx}$
>- `np.arctan2()` --> $gradAz = arctang(\frac{gy}{gx})$ in radians
>- $gradAz = gradAz \cdot \frac{180}{\pi}$ in degrees, from -180 to 180
>- $gradAz = gradAz + 180$  in degrees, from 0 to 360

```python
gradAz = (np.arctan2(gy,gx) * 180 / np.pi)+180
plt.imshow(gradAz);
```

```python
gradAz.min(),gradAz.max()
```

```python
plt.hist(gradAz.ravel(),bins=361);
```

## Gradient azimuth transformation
> - gradient azimuth transformation into 8 directions (every 45 degrees)

```python
step=45
n = int(360/step)
angles = np.array([step*i for i in range(0,n+1)])
print(angles, step/2, sep='\n\n')
```

```python
gradAz1 = np.zeros_like(gradAz)
for x in angles:
    print(f'{x}: {x-step/2}, {x+step/2}')
    idx = (gradAz >= x-step/2) & (gradAz < x+step/2)
    gradAz1[idx] = x
    
mymod.myshow([gradAz,gradAz1],1,2,4)
```

```python
val, nc = np.unique(gradAz1,return_counts=1)
plt.bar(val,nc);
plt.xticks(angles);
```

## Selecting pixels in the edge

with the highest values

```python
#rows,cols = grad.shape
thin = grad.copy()
pad = np.pad(grad,1)
rows,cols = pad.shape
#i=0
for r in range(1,rows-1):
    for c in range(1,cols-1):
        
        window = pad[r-1:r+2,c-1:c+2]
        az = gradAz1[r-1,c-1]
        
        if az == 0 or az ==180 or az==360:             # 0/180 degrees
            pix = max(window[[1,1],[0,2]])
            
        elif az == 45 or az == 45+180:             # 45/225 degrees
            pix = max(window[[2,0],[0,2]])
            
        elif az == 90 or az == 90+180:             # 90/270 degrees
            pix = max(window[[0,2],[1,1]])
            
        elif az == 135 or az == 135+180:             # 135/315 degrees
            #i+=1
            pix = max(window[[0,2],[0,2]])            
            #if i < 6:
                #print(f'i: {i}\n{window}\n\n{window[[0,2],[0,2]]}\np1: {pix}\n\n')
        
        #print(f'{window}\n\naz: {az}, pix: {pix}\n\n')
        
        if window[1,1] < pix:
            #print(f'window[1,1]: {window[1,1]}, pix: {pix} - {window[1,1]<pix}')
            thin[r-1,c-1] = 0
```

# Double thresholding

>*Aim*:   
division of all pixels into 3 groups:
>- pixels that definitely define the edge
>- pixels that definitely do not define an edge
>- uncertain pixels that cannot be omitted at this stage 

```python
plt.hist(thin.ravel(),bins=50);
```

```python
def fthresh(t1,t2):
    ar = thin.copy()
    t1 = ar.max() * t1
    t2 = ar.max() * t2
    
    idx1 = thin <= t1
    idx2 = thin > t2
    idx3 = (thin > t1) & (thin <= t2)
    
    ar[idx1] = 0
    ar[idx2] = 255
    ar[idx3] = 100
    
    mymod.myshow([thin,ar],1,2,6)
    
interact(fthresh, t1=(0,0.1,0.01),t2=(0.1,1,0.05))
```

```python
def fthresh1(t1,t2):
    ar = thin.copy()
    t1 = ar.max() * t1
    t2 = ar.max() * t2
    idx1 = thin <= t1
    idx2 = thin > t2
    idx3 = (thin > t1) & (thin <= t2)
    
    ar[idx1] = 0
    ar[idx2] = 255
    ar[idx3] = 100
    return ar
```

```python
reclas = fthresh1(0.05,0.15)
plt.imshow(reclas);
```

# Edge Tracking by Hysteresis

Hysteresis processing uses uncertain pixels to build edges.
>- an uncertain pixel is classified as an edge only when it is in contact with another pixel belonging to the edge
>- otherwise it is classified as background 

```python
np.unique(reclas)
```

```python
res = np.zeros_like(reclas,dtype=np.uint8)
pad = np.pad(reclas,1)
rows,cols = pad.shape
maxpix = reclas.max()

for r in range(1,rows-1):
    for c in range(1,cols-1):
        
        window = pad[r-1:r+2,c-1:c+2]
        
        if window.max() == maxpix:
            #print(f'window.max(): {window.max()}, maxpix: {maxpix} - {window.max()==maxpix}')
            res[r-1,c-1] = 255
        else:
            res[r-1,c-1] = 0

            
mymod.myshow([grad,reclas,res],1,3,5)
```
