```python
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
import geojson
from shapely import geometry
from sentinelsat import SentinelAPI
from datetime import date
import utm
from ipyleaflet import Map, Marker, basemaps, basemap_to_tiles
from ipyleaflet import Polygon as ipylPolygon
```

```python
geolocator.geocode?
```

```python
address = {'country':'Poland','city':'Krakow'}
geolocator = Nominatim(user_agent="Pio")
```

```python
location = geolocator.geocode(address)
type(location)
```

```python
location.latitude, location.longitude
```

3. utm module - conversion of geographic coordinates

```python
x, y, zone, zoneLetter = utm.from_latlon(location.latitude, location.longitude)
print(x,y)
```

```python
offset = 15*1000

upl = (x-offset, y+offset)
upr = (x+offset, y+offset) #upper right

dwr = (x+offset, y-offset)
dwl = (x-offset, y-offset) #down left
```

```python
utm.to_latlon?
```

```python
geoCord = []
for p in [upl,upr,dwr,dwl]:
    geoCord.append(utm.to_latlon(p[0],p[1],zone,zoneLetter))
    
geoCord
```

4. ipyleaflet - display the map

```python
latlon = (location.latitude,location.longitude)
 
m = Map(
    basemap=basemap_to_tiles(basemaps.OpenTopoMap, "2017-08-08"),
    center=(latlon[0],latlon[1]),
    zoom=10
)
 
m.add_layer(Marker(location=latlon))
area = ipylPolygon(locations=geoCord,color='red',fill_opacity= 0.1,weight=2)
m.add_layer(area)
m
```

5. geojson

```python
cords = geoCord.copy()
cords.append(geoCord[0])

polyGeojson = geojson.Polygon(cords)
print(polyGeojson, '\n\n', type(polyGeojson),'\n')
```

6. shapely

```python
polyShapely = geometry.asPolygon(cords)
polyShapely
```

```python
polyShapely.to_wkt()
```
