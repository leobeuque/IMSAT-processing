# 1. Histogram stretching

Aim: to change the contrast of the image

Method: linear histogram stretching

Contrast: the difference between the maximum and minimum image value

Equation: $ y = ax + b $, where:

   - x: curren image values
   - y: new image values
   - a, b: transformation coefficients 

```python
tmp = dt['sent'].copy()
print(tmp.shape)

f,ax = plt.subplots(1,3,figsize=(12,4))
for i in range(tmp.shape[-1]):
    ax[i].hist(tmp[...,i].ravel(),bins=256)
```

```python
b1 = tmp[...,0].copy()
b1.min(),b1.max()
```

```python
# percentile
a = np.arange(101)
print(a)
p = 1
x1, x2 = np.percentile(a,[p,100-p])
x1,x2
```

```python
# y = ax + b
# a, b ?
# x1 --> ny1, x2 --> ny2
x1, x2 = np.percentile(b1,[p,100-p])
x1,x2
(x1,0), (x2,255)
a,b = np.polyfit([x1,x2],[0,255],1)
a,b
```

```python
y[y>255].max()
```

```python
x = range(256)
y = x*a + b
y1 = y.copy()

y[y<0] = 0
y[y>255] = 255

f = plt.figure(figsize=(5,5))
plt.plot(x,y,label='real new pixels')
plt.plot(x,y1,'--k',label='new theoretical pixels from formula y=ax+b')

xpoints = [x1,x2]
ypoints = [a*j + b for j in xpoints]

plt.plot(xpoints,ypoints,'or',label='thresholds: min and max')

plt.xlabel('old image')
plt.ylabel('new image')

plt.xticks(list(range(0,256,25)));
plt.yticks(np.arange(-50,350,25));
plt.ylim(-50,350)

plt.grid()
plt.legend(loc='center',bbox_to_anchor=(0.5, 0.5, 0, 1.22));
```

```python
nb1 = np.floor(b1.copy() * a + b)
nb1[nb1 < 0] = 0
nb1[nb1 > 255] = 255
np.unique(nb1)

f, ax = plt.subplots(1,2,figsize=(16,6))
ax[0].imshow(b1,'gray')
ax[1].imshow(nb1,'gray')
```

```python
def stret(p):
    '''p: int (0-100), percentile'''
    ar = b1.copy()
    x1, x2 = np.percentile(ar,[p,100-p])
    print(x1,x2)

    a,b = np.polyfit([x1,x2],[0,255],1)
    ar1 = np.floor(ar.copy() * a + b)
    ar1 [ar1  < 0] = 0
    ar1 [ar1  > 255] = 255
    
    f, ax = plt.subplots(1,2,figsize=(16,6))
    ax[0].imshow(ar,'gray')
    ax[1].imshow(ar1,'gray')
    
interact(stret, p=(0,20))
```

```python
def stret1(p):
    '''p: int (0-100), percentile'''
    ar = dt['sent'].copy()
    ar1 = ar.copy()
    
    for i in range(ar.shape[-1]):
        band = ar[...,i].copy()
        x1, x2 = np.percentile(band,[p,100-p])

        a,b = np.polyfit([x1,x2],[0,255],1)
        band = np.floor(band * a + b)
        band[band<0] = 0
        band[band>255] = 255
        ar1[...,i] = band.copy()
        
    
    f, ax = plt.subplots(2,2,figsize=(10,10))
    ax = ax.ravel()
    ax[0].imshow(ar,'gray')
    ax[1].imshow(ar1,'gray')
    ax[2].hist(ar.ravel(),bins=256)
    ax[3].hist(ar1.ravel(),bins=256)
```

```python
interact(stret1, p=(0,20))
```

# Histogram equalization


$$im_{e_{i,j}} = (L - 1) \sum_{0,0}^{im_{i,j}} p_n$$

    
Where:
 - $im, im_e$ - images: oryginal, after equalize
 - $i,j$ - number of rows and columns. A pair of numbers (i, j) indicate a specific pixel in the image 
 - $L$ - radiometric resolution (the number of bits to record information in the image, usually 8 bits = 256 values) 
 - $n$ - the frequency of occurrence of values in the image 
 - $p_n$ - the probability of n-value in the image, i.e. the frequency of occurrence in the image divided by the number of all values


### Probability $p_n$
>- `n` frequency of occurrence e.g. `np.unique()` with `return_counts` argument set to `True`
>- the number of total values (pixels) in an image is its size: `im.size`
>- $p_n = \frac{n}{im.size}$


### `CDF` function
>cumulative distributive function

$$cdf = \sum_{i,j}^{im_{i,j}} p_n$$


### `L-1`
>- $L = 2^8 = 256$ - for 8 bits
>- $L - 1 = 255$

---

## Steps

- calculate the unique values of pixels in the image `pix` and their frequencies `n`
- calculate the probabilities $p_n$
- calculate `cdf`
- calculate new pixel values of `npix`
- map new values (`npix`) to old ones (`pix`) throughout the image

```python
np.random.seed(10)
at =np.random.randint(0,20,size=(4,4))
at
```

```python
np.unique(at,return_counts=True)
```

```python
b1 = dt['sent'][...,0].copy()
pix, n = np.unique(b1,return_counts=True)
plt.bar(pix,n);
```

```python
pn = n / b1.size
pn.shape, pn[:10]
```

```python
cdf = np.cumsum(pn)
plt.plot(pix,cdf);
```

```python
npix = 255*cdf
npix = ((npix - npix.min()) / (npix.max()-npix.min())) * 255
npix = np.floor(npix).astype(np.uint8)
np.unique(npix)
```

```python
nb = b1.copy()
for i,x in enumerate(pix):
    idx = b1 == x
    newx = npix[i]
    nb[idx] = newx
```
